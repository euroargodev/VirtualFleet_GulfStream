```python
get_ano = lambda x: {'name': "%s vs %s" % (x['exp']['name'], x['ctl']['name']), 'region': x['exp']['region']}
def get_res(x, exp): 
    x['exp'] = exp
    x['ano'] = get_ano(x)
    return x

figures = {}

figures[results['obs']['name']] = get_obs_figure_list(results)
figures[results['ctl']['name']] = get_ctl_figure_list(results)

for exp in exp_list:
    res = get_res(results.copy(), exp)
    l = get_exp_figure_list(res, 'exp')
    figures[exp['name']] = l
    
df = pd.DataFrame(figures)
df    
```

```python
def format_link(link):
    root = "https://raw.githubusercontent.com/euroargodev/VirtualFleet_GulfStream/main"
    if link:
        return "![](%s/%s)" % (root, link.replace("#","%23"))
    else:
        return ""    

MDlines = []    
MDlines.append("|%25s|%s|" % (" ", "|".join(['{0:{1}s}'.format(k, col_size) for k in headers])))
MDlines.append("|%24s:|:%s|" % ("-"*24, "|:".join(["-"*(col_size-1) for k in headers])))
MDlines.append("|%25s|%s|" % ("Definition", "|".join(['{0:{1}s}'.format(k, col_size) for k in subhead])))
for ii, row in df.iterrows():
    cols = row.values
    # print("|%25s|%s|" % ("Figure %i" % (ii+1), "|".join(['{0:{1}s}'.format(k, col_size) for k in cols])))
    MDlines.append("|%25s|%s|" % ("Figure %i" % (ii+1), "|".join([format_link(k) for k in cols])))
MDlines    
```
